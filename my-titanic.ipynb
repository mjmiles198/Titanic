# Feature Extraction

```python
gender=pd.concat([train_df['Sex'],test_df['Sex']],axis=0)
#fare=pd.concat([train_df['Fare'],test_df['Sex']],axis=0)
```

```python
def impute_age(cols):
    Age = cols[0]
    Pclass = cols[1]
    
    if pd.isnull(Age):
        
        if Pclass == 1:
            return 37
        elif Pclass == 2:
            return 29
        else:
            return 24
    else:
        return Age
```

```python
train_df['Age'] = train_df[['Age', 'Pclass']].apply(impute_age, axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
gender_transfomer=LabelEncoder()
#gender_transfomer.fit(gender)
```

```python
train_df['Sex']=gender_transfomer.fit_transform(train_df['Sex'])
test_df['Sex'] =gender_transfomer.transform(test_df['Sex'])
```

```python
train_df.head()
```

```python
ms=MinMaxScaler()
train_df['Fare']=ms.fit_transform(train_df[['Fare']])
test_df['Fare'] =ms.transform(test_df[['Fare']])
train_df.head()
```

```python
train_df['Age']=ms.fit_transform(train_df[['Age']])
test_df['Age'] =ms.transform(test_df[['Age']])
```

```python
train_df.dropna()
```

```python
y=train_df['Survived']
train_df.drop(['Survived','Ticket','Embarked','SibSp','Name','Cabin'],axis=1, inplace=True)
train_df.head()              
```

```python
train_df.isna().sum()
```

```python
train_df.dropna()
```

```python
X_train, X_test, y_train, y_test=train_test_split(train_df,y,test_size=0.3)
```
